```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

```python
school_data_complete.head()
```

```python
#Calculate the total number of schools
schools_total = len(school_data_complete["school_name"].unique())
schools_total
```

```python
#Calculate the total number of students
students_total= school_data_complete["Student ID"].count()
students_total 
```

```python
#Calculate the total budget
total_budget = school_data_complete["budget"].sum()
total_budget 
```

```python
#Calculate the average math score
average_math = school_data_complete["math_score"].mean()
average_math

#Calculate the average reading score
average_reading = school_data_complete["reading_score"].mean()
average_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
count_math=school_data_complete[(school_data_complete['math_score'] >= 70)].count()['math_score']
percentage_math=count_math/students_total*100

#Calculate the percentage of students with a passing reading score (70 or greater)
count_reading=school_data_complete[(school_data_complete['reading_score'] >= 70)].count()['reading_score']
percentage_reading=count_reading/students_total*100

#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
count_math_read=school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >=70)].count()['student_name']
percentage_overall_passing = count_math_read/students_total*100
```

```python
#Create a dataframe 
district_summary_df = pd.DataFrame({"total_schools": [schools_total],
                              "total_students": students_total,
                              "total_budget": total_budget,
                              "average_mat_score": average_math,
                              "average_reading_score": average_reading, 
                             "percentage_passed_math": percentage_math,
                             "percentage_passed_reading": percentage_reading,
                             "percentage_overall_passing": percentage_overall_passing})

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Total Students - for each school 
school_student_count = school_data_complete["school_name"].value_counts()

# Total School Budget - for each school 
total_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]

# Per student budget - total budget/total students 
per_student_budget = total_budget/students_total
```

```python
# Average Math Score 
average_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]

# Average Reading Score 
average_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
#School Percentage Passing Math 
school_passing_math=school_data_complete.loc[school_data_complete["math_score"]>=70]
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / school_student_count * 100

#School Percentage Passing Reading
school_passing_read=school_data_complete.loc[school_data_complete["reading_score"]>=70]
per_school_passing_read = school_passing_read.groupby(["school_name"]).count()["student_name"] / school_student_count * 100

# School Overall Passing (passing math and reading)
school_passing_math_read=school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
per_school_passing_math_read = school_passing_math_read.groupby(["school_name"]).count()["student_name"] / school_student_count * 100

# Create a dataframe  
school_summary_df = pd.DataFrame({"total_students": school_student_count,
                               "total_budget": total_budget, 
                               "per_student_budget": per_student_budget, 
                                "average_math_score": average_math, 
                                "average_reading_score": average_reading,
                                "percentage_passed_math": per_school_passing_math,
                                "percentage_passed_reading": per_school_passing_read, 
                                "percentage_overall_passing": per_school_passing_math_read})

school_summary_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Sorted by % overall passing  
top_performing_schools = school_summary_df.sort_values("percentage_overall_passing", ascending=False)
top_performing_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
# Sorted by percent overall passing 
top_performing_schools = school_summary_df.sort_values("percentage_overall_passing")
top_performing_schools.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. 
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
#9th 
ninth_grade = school_data_complete.loc[school_data_complete["grade"]== "9th"] 
ninth_grade_by_school = ninth_grade.groupby("school_name")
ninth_grade_math_average = ninth_grade_by_school.mean()["math_score"]

#10th 
tenth_grade = school_data_complete.loc[school_data_complete["grade"]== "10th"] 
tenth_grade_by_school = tenth_grade.groupby("school_name")
tenth_grade_math_average = tenth_grade_by_school.mean()["math_score"]

#11th
eleventh_grade = school_data_complete.loc[school_data_complete["grade"]== "11th"] 
eleventh_grade_by_school = eleventh_grade.groupby("school_name")
eleventh_grade__math_average = eleventh_grade_by_school.mean()["math_score"]

#12th
twelfth_grade = school_data_complete.loc[school_data_complete["grade"]== "12th"] 
twelfth_grade_by_school = eleventh_grade.groupby("school_name")
twelfth_grade__math_average = eleventh_grade_by_school.mean()["math_score"]


#create summary dataframe
math_average_df = pd.DataFrame ({"9th": ninth_grade_math_average,
                    "10th": tenth_grade_math_average, 
                    "11th": eleventh_grade__math_average, 
                    "12th": twelfth_grade__math_average})                 
math_average_df
```
